When we join the customers, rentals, and movies tables into one large table, each row represents a rental transaction. However, this creates redundancy:Customer information name, email, city is repeated for every rental they make. Movie information title, genre, rating is repeated every time the movie is rented.
This redundancy can cause problems:

If a customer updates their email, we would need to update multiple rows.

If a movie’s rating changes, it must be updated everywhere it appears.

Deleting rentals could accidentally remove important customer or movie data.

The normalized design avoids these issues by storing customers, movies, and rentals in separate tables and linking them using IDs. This reduces duplication, improves data integrity, and makes updates safer and more efficient.

```python
customers_docs = []

for index, cust in customers.iterrows():

    cust_rentals = rentals[rentals["customer_id"] == cust["customer_id"]]
    
    rental_list = []
    
    for index, rent in cust_rentals.iterrows():

        movie = movies[movies["movie_id"] == rent["movie_id"]].iloc[0]
        
        rental_doc = {
            "rental_id": rent["rental_id"],
            "rental_date": rent["rental_date"],
            "return_date": rent["return_date"],
            "price": rent["price"],
            "movie": {
                "movie_id": movie["movie_id"],
                "title": movie["title"],
                "genre": movie["genre"],
                "release_year": movie["release_year"],
                "rating": movie["rating"]
            }
        }
        
        rental_list.append(rental_doc)
    
    customer_doc = {
        "customer_id": cust["customer_id"],
        "name": cust["name"],
        "email": cust["email"],
        "city": cust["city"],
        "signup_date": cust["signup_date"],
        "rentals": rental_list
    }
    
    customers_docs.append(customer_doc)

print(customers_docs)
```

```python
# 1) Find all movies rented by a specific customer.
customers = {}
for offer in customers_docs:
    for rental in offer["rentals"]:
        if offer["name"] in customers:
            if rental["movie"]["title"] not in customers[offer["name"]]:
                customers[offer["name"]].append(rental["movie"]["title"])
        else:
            customers[offer["name"]] = []
            customers[offer["name"]].append(rental["movie"]["title"])
print(customers)

```

```python
# 2) Find the top 5 most-rented movies.
movie_rent = {}
for offer in customers_docs:
    for rental in offer["rentals"]:
        if rental["movie"]["title"] in movie_rent.keys():
            movie_rent[rental["movie"]["title"]] += 1 
        else:
            movie_rent[rental["movie"]["title"]] = 1 

top5 = sorted(movie_rent.items(), key=lambda x: x[1], reverse=True)[:5]
top5
```

```python
# 3) Compute total revenue per genre.
revenue_per_genre = {}

for offer in customers_docs:
    for rental in offer["rentals"]:
        genre = rental["movie"]["genre"]
        price = rental["price"]
        
        if genre in revenue_per_genre:
            revenue_per_genre[genre] += price
        else:
            revenue_per_genre[genre] = price

revenue_per_genre
```

```python
# 4) Find customers who have never rented a movie rated below 3.

good_customers = []

for offer in customers_docs:
    has_low_rating = False
    
    for rental in offer["rentals"]:
        if rental["movie"]["rating"] < 3:
            has_low_rating = True
            break
    
    if not has_low_rating:
        good_customers.append(offer["name"])

good_customers
```

In our exercises, queries in the document model were easier when retrieving all information for a single customer because the data is nested and no joins are needed. However, analytical queries like counting top movies or filtering customers by ratings were more complex, requiring loops and manual checks. In contrast, the relational model made aggregation queries much simpler using SQL features like GROUP BY, SUM, and HAVING. Joins across tables can be tricky at first, but once set up, they are very efficient for analytics. The document model is best for hierarchical or flexible data, while the relational model is best for structured data and reporting tasks. Overall, document databases are easier for application-focused queries, whereas relational databases are better for large-scale analysis and summaries.

```python
nodes = {
    "c001": {"type": "customer", "name": "Alice", "city": "Baku"},
    "c002": {"type": "customer", "name": "Bob", "city": "Ganja"},
    "c003": {"type": "customer", "name": "Charlie", "city": "Baku"},
    
    "m001": {"type": "movie", "title": "Inception"},
    "m002": {"type": "movie", "title": "Titanic"},
    "m003": {"type": "movie", "title": "Avengers"},
    
    "g001": {"type": "genre", "name": "Sci-Fi"},
    "g002": {"type": "genre", "name": "Romance"},
    "g003": {"type": "genre", "name": "Action"},
}

edges = [
    ("c001", "rented", "m001"),
    ("c001", "rented", "m003"),
    ("c002", "rented", "m002"),
    ("c003", "rented", "m001"),
    
    ("m001", "belongs_to", "g001"),
    ("m002", "belongs_to", "g002"),
    ("m003", "belongs_to", "g003"),
]
```

```python
def movies_rented_by(customer_id):
    return [
        nodes[target]["title"]
        for source, relation, target in edges
        if source == customer_id and relation == "rented"
    ]

print(movies_rented_by("c001"))
```

```python
from collections import Counter

def preferred_genres(customer_id):
    genres = []
    
    rented_movies = [
        target for source, relation, target in edges
        if source == customer_id and relation == "rented"
    ]
    
    for movie in rented_movies:
        for source, relation, target in edges:
            if source == movie and relation == "belongs_to":
                genres.append(nodes[target]["name"])
    
    return Counter(genres)

print(preferred_genres("c001"))
```

```python
def customers_with_shared_genre(customer_id):
    shared_customers = set()
    
    customer_genres = set(preferred_genres(customer_id).keys())
    
    for other_id, data in nodes.items():
        if data["type"] == "customer" and other_id != customer_id:
            other_genres = set(preferred_genres(other_id).keys())
            if customer_genres.intersection(other_genres):
                shared_customers.add(data["name"])
    
    return shared_customers

print(customers_with_shared_genre("c001"))
```

```python
nodes["m004"] = {"type": "movie", "title": "The Notebook"}

edges.append(("m004", "belongs_to", "g002"))
def recommend_movies(customer_id):
    rented_movies = set(
        target for source, relation, target in edges
        if source == customer_id and relation == "rented"
    )
    
    customer_genres = set(preferred_genres(customer_id).keys())
    
    recommendations = []
    
    for movie_id, data in nodes.items():
        if data["type"] == "movie" and movie_id not in rented_movies:
            
            for source, relation, target in edges:
                if source == movie_id and relation == "belongs_to":
                    if nodes[target]["name"] in customer_genres:
                        recommendations.append(data["title"])
    
    return recommendations

print(recommend_movies("c002"))
```

The recommendation query requires moving through several relationships: from a customer to the movies they rented, then to the genres of those movies, and finally to other movies in those same genres that the customer has not rented yet. In a relational database, this would require multiple joins between tables, which can make the query longer and harder to read. In a document model, we would usually need nested loops or repeated filtering to follow these relationships. In contrast, graph databases store relationships directly, making it easy to traverse connections step by step, which is why they are well suited for recommendation systems.

OLTP workloads usually involve inserting, updating, or retrieving a single transaction. These operations access all fields of one record, so row-major storage is efficient because all values of that row are stored together in memory. This makes reading or updating a full transaction fast.

OLAP workloads focus on analyzing large datasets using operations like sums, averages, and counts. These queries often use only a few columns but across many rows. Column-major storage groups values from the same column together, which makes scanning and aggregation faster. Therefore, OLTP systems work best with row-based storage, while OLAP systems benefit from column-based storage for analytics.
